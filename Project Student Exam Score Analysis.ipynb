```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dropping Unnamed Column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (5,5))
ax=sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# From the above chart we have analysed that the number of females in the data is more than the number of males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print (gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's score")
plt.show()
```

```python
# From the above chart we have concluded that the education of Parents have good impact on their scores.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot = True)
plt.title("Relationship between Parent's Marital Status and Student's score")
plt.show()
```

```python
# From the above chart we have concluded that there is no/negligible impact on student's score due to their parent's marital status.
```

# Distribution of Ethnic Group

```python
l=["group A","group B","group C","group D","group E"]


groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()

```
